```python
# Plot the data
data.plot(x='price', y='Area_in_Marla', s = 32, alpha = 0.5, kind = 'scatter')
```

```python
# Making a bar plot for the number of baths
baths_count = data['baths'].value_counts()
baths_count = baths_count.sort_index()
baths_count.plot(kind = 'bar')
plt.title('Number of baths')
plt.xlabel('Number of baths')
plt.ylabel('Count')

plt.show()
```

```python
# Making a bar plot for the number of bedrooms
bedrooms_count = data['bedrooms'].value_counts()
bedrooms_count = bedrooms_count.sort_index()
bedrooms_count.plot(kind = 'bar')
plt.title('Number of bedrooms')
plt.xlabel('Number of bedrooms')
plt.ylabel('Count')

plt.show()
```
